### Вводная лекция

```python
import numpy as np
import matplotlib.pyplot as plt

def dot_list(x: list, y: list):
    res = 0
    for i in range(len(x)):
        res += x[i] * y[i]
    return res

def dot_np(x, y):
    return np.dot(x, y)

%timeit dot_list(list(range(1000)), list(range(1000)))    
```

```python
%timeit dot_np(np.arange(1000), np.arange(1000))    
```
